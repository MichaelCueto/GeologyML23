```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
def Statistics(mask,df):
    for i in mask:
        print('----------------------------')
        print(df[i].describe().transpose())
        random_color = np.random.rand(3,)
        
        sns.set_theme()
        sns.lineplot(data=df[i], color=random_color)
        plt.title(f'Line Plot: {i}')
        plt.show()
        sns.histplot(data=df[i], kde=True, color=random_color)
        plt.title(f'Histogram Plot: {i}')
        plt.show()
```

***Métricas y Estadísticas de las variables usadas en el archivo de validacion de Dureza***

```python
mask = ['Cu%', 'Mo%','AgPPM', 'AuPPM', 'Al%', 'Ca%', 'Fe%', 'Mg%', 'Na%', 'S%', 'BWIP', 'SPIP']
root = '../03_BBDD/bwilt23.3.xlsx'
df = pd.read_excel(root)
Statistics(mask,df)
```

***Métricas y Estadísticas de las variables usadas en la data de Entrenamiento de Dureza***

```python
mask = ['Cu%', 'Mo%','AgPPM', 'AuPPM', 'Al%', 'Ca%', 'Fe%', 'Mg%', 'Na%', 'S%','BWI','SPI']
root = '../03_BBDD/data_dureza_train.xlsx'
df = pd.read_excel(root)
Statistics(mask,df)
```

***Correlación de variables según la data de entrenamiento***

```python
correlation = df[mask].corr().stack().reset_index(name='correlation')
correlation_ = df[mask].corr()
sns.relplot(correlation, x="level_0", y="level_1",hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor="0.1",
    height=8, sizes=(30, 150), size_norm=(-.2, .8))
plt.show()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_,annot=True, fmt=".2f",linewidth=.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Assuming 'df' is your DataFrame with the columns 'X' and 'Y'
# Replace df with your own data
root = '../03_BBDD/data_dureza_train.xlsx'
df = pd.read_excel(root)
# Select the columns X and Y from the DataFrame
X = df[['X', 'Y']].values

# Apply KMeans for segmentation
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(X)

# Create a 2D scatter plot for X and Y with colors based on KMeans segmentation
plt.figure(figsize=(10, 8))
for i in range(n_clusters):
    cluster_points = X[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i + 1}')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('KMeans Segmentation - X and Y')

# Show the legend and the plot
plt.legend()
plt.show()
```
